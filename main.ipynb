```python
counts = {1: 0, 2: 0, 3: 0}

# open the file
with open('./dataset/cls3_train.txt', 'r') as f:
    # iterate over each line
    for line in f:
        # split the line by spaces
        parts = line.split()
        # get the last part as an integer
        try:
            last_int = int(parts[-2])
            # if the integer is in the desired range, increment the corresponding count
            if last_int in counts:
                counts[last_int] += 1
        except ValueError:
            # ignore lines that do not end with an integer
            pass

# print the results
for key, value in counts.items():
    print(f"Total count for birdas {key}: {value}")
```
